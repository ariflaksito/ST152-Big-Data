## Melihat Video paling banyak like

```python
df_like = df.sort_values(by=['likeCount'], ascending=False)
df_like.head()
```

```python
df_like_stats = df_like.iloc[:10]
df_like_stats
```

```python
plt.figure(figsize=(12,7))
plt.bar(df_like_stats['title'],df_like_stats['likeCount'])
plt.title('Jumlah like di video terbanyak')
plt.xlabel('Judul Video')
plt.ylabel('Jumlah like')
plt.xticks(rotation=90, ha='right')

plt.show()
```
